```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("StudentsPerformance.csv")
```

```python
sns.distplot(df['math score'])
```

```python
sns.distplot(df['reading score'])
```

```python
sns.distplot(df['writing score'])
```

```python
sns.distplot(df['math score'],kde=False,hist_kws=dict(edgecolor="k", linewidth=1))
```

```python
sns.kdeplot(df['math score'], shade=True)
sns.kdeplot(df['reading score'], shade=True)
```

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

```python
sns.pairplot(df)
```

```python
sns.countplot(x="gender", data=df,hue='lunch')
```

```python
sns.countplot(x="gender", data=df,hue='race/ethnicity')
```

```python
sns.countplot(df["gender"])
print (df['gender'].value_counts())
```

```python
f,axes = plt.subplots(figsize=(10,10),sharex=True)
sns.scatterplot(x='writing score',y='math score',data=df,hue=df['gender'])
```

```python
sns.countplot(df['parental level of education'])
plt.xticks(rotation=30)
print (df['parental level of education'].value_counts())
```
